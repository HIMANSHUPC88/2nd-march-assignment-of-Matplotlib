# Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

```python
# Scatter plot is a type of visulization which uses points to represents values for two different variable.helping to illustrate the relationship between those variable.
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y, c = 'r' ,  marker = 'X' , alpha = .4)
plt.xlabel(" X Axis ")
plt.ylabel(" Y Axis ")
plt.show()
```

```python
#subplot() function is used for visulize more then one plot in a single plot.it takes 3 arguments. 
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3= np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title("One")

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title("Two")

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title("Three")

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title("Four")


plt.show()
```

# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot. 

# import numpy as np

# company = np.array(["Apple", "Microsoft", "Google", "AMD"])

# profit = np.array([3000, 8000, 1000, 10000])

```python
English - Bar plot is a type of visulization. which is used to represent or compare the value of two different categories or groups. 
Hinglish - Bar plot ek type of visulization hota hai jo Different Category ya Different group ki values ko compare ya represent karta hai.
```

```python
import numpy as np
import matplotlib.pyplot as plt
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit , alpha = .5 )
plt.xlabel("NAME OF COMPANY")
plt.ylabel("PROFIT")
plt.show()
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.barh(company,profit , alpha = .5  )
plt.xlabel("NAME OF COMPANY")
plt.ylabel("PROFIT")
plt.show()
```
